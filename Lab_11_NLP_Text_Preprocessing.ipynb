## 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

import re
import string

# Download required NLTK data
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('vader_lexicon', quiet=True)

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

print("All NLP libraries loaded successfully.")
```

## 2. Prepare Sample Text Data

```python
# Sample corpus for demonstration
corpus = [
    "Natural Language Processing (NLP) is a subfield of artificial intelligence that focuses on the interaction between computers and human language. It enables machines to read, understand, and derive meaning from human languages.",
    "Text preprocessing is a crucial step in NLP. It involves cleaning and transforming raw text data into a format suitable for analysis. Common techniques include tokenization, stop-word removal, and stemming.",
    "Machine learning algorithms can be applied to NLP tasks such as sentiment analysis, text classification, named entity recognition, and machine translation. These applications have revolutionized how we interact with technology.",
    "Deep learning models like transformers have significantly advanced the field of NLP. Models such as BERT, GPT, and T5 achieve state-of-the-art results on various language understanding benchmarks.",
    "The Seeds dataset contains measurements of wheat kernels. While not a text dataset, we can demonstrate NLP techniques by analyzing descriptive text about agricultural data and crop classification methods."
]

print(f"Sample corpus loaded with {len(corpus)} documents.")
for i, doc in enumerate(corpus, 1):
    print(f"\nDocument {i}:")
    print(doc)
```

## 3. Tokenization

```python
# --- Sentence Tokenization ---
print("=== SENTENCE TOKENIZATION ===")
sample_text = corpus[0]
sentences = sent_tokenize(sample_text)
print(f"\nDocument 1 → {len(sentences)} sentences:")
for i, s in enumerate(sentences, 1):
    print(f"  [{i}] {s}")

# --- Word Tokenization ---
print("\n=== WORD TOKENIZATION ===")
word_tokens = word_tokenize(sample_text)
print(f"\nDocument 1 → {len(word_tokens)} tokens:")
print(f"  {word_tokens}")
```

## 4. Text Cleaning & Normalization

```python
def clean_text(text):
    """Apply basic text cleaning operations."""
    # Lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

print("=== TEXT CLEANING PIPELINE ===")
sample = corpus[0]
print(f"\nOriginal:\n  {sample}")

# Step by step
step1 = sample.lower()
print(f"\nAfter lowercasing:\n  {step1}")

step2 = step1.translate(str.maketrans('', '', string.punctuation))
print(f"\nAfter removing punctuation:\n  {step2}")

step3 = re.sub(r'\s+', ' ', step2).strip()
print(f"\nAfter removing extra whitespace:\n  {step3}")
```

## 5. Stop-Word Removal

```python
print("=== STOP-WORD REMOVAL ===")

stop_words = set(stopwords.words('english'))
print(f"\nTotal English stop words in NLTK: {len(stop_words)}")
print(f"Sample stop words: {list(stop_words)[:10]}")

# Clean and tokenize
cleaned = clean_text(corpus[0])
tokens = cleaned.split()

# Remove stop words
filtered_tokens = [w for w in tokens if w not in stop_words]

print(f"\nOriginal tokens ({len(tokens)}): {tokens[:30]}")
print(f"\nAfter stop-word removal ({len(filtered_tokens)}): {filtered_tokens}")
print(f"\nRemoved {len(tokens) - len(filtered_tokens)} stop words ({(len(tokens) - len(filtered_tokens))/len(tokens)*100:.1f}% reduction)")
```

## 6. Stemming

```python
print("=== STEMMING (Porter Stemmer) ===")

stemmer = PorterStemmer()

# Demonstrate stemming on individual words
demo_words = ['processing', 'processes', 'processed', 'intelligence', 'intelligent',
              'running', 'runner', 'classification', 'classified',
              'measurements', 'measuring', 'languages', 'understanding',
              'revolutionized', 'transforming']

print(f"\n{'Word':<20} → {'Stem':<20}")
print("-" * 44)
for word in demo_words:
    print(f"{word:<20} → {stemmer.stem(word):<20}")

# Apply to filtered tokens
stemmed_tokens = [stemmer.stem(w) for w in filtered_tokens]
print(f"\nApplied to Document 1:")
print(f"  Original:  {filtered_tokens}")
print(f"  Stemmed:   {stemmed_tokens}")
```

## 7. Lemmatization

```python
print("=== LEMMATIZATION (WordNet) ===")

lemmatizer = WordNetLemmatizer()

demo_words_lemma = ['processing', 'processes', 'processed', 'intelligence', 'intelligent',
                    'running', 'runner', 'languages', 'understanding',
                    'measurements', 'mice', 'better', 'corpora']

print(f"\n{'Word':<20} → {'Lemma':<20}")
print("-" * 44)
for word in demo_words_lemma:
    print(f"{word:<20} → {lemmatizer.lemmatize(word):<20}")

# Compare Stemming vs Lemmatization
print("\n=== STEMMING vs. LEMMATIZATION ===")
compare_words = ['processing', 'languages', 'intelligence', 'running', 'measurements']
print(f"\n{'Word':<20}{'Stem':<20}{'Lemma':<20}")
print("-" * 60)
for word in compare_words:
    print(f"{word:<20}{stemmer.stem(word):<20}{lemmatizer.lemmatize(word):<20}")
```

## 8. POS Tagging

```python
print("=== PART-OF-SPEECH (POS) TAGGING ===")

sample_tokens = word_tokenize(corpus[0])
pos_tags = pos_tag(sample_tokens)

# POS tag descriptions
tag_descriptions = {
    'NN': 'Noun, singular or mass', 'NNS': 'Noun, plural',
    'NNP': 'Proper noun, singular', 'VB': 'Verb, base form',
    'VBZ': 'Verb, 3rd person singular present', 'VBG': 'Verb, gerund',
    'JJ': 'Adjective', 'RB': 'Adverb', 'IN': 'Preposition or subordinating conjunction',
    'DT': 'Determiner', 'WDT': 'Wh-determiner', 'TO': 'to',
    'CC': 'Coordinating conjunction', 'PRP': 'Personal pronoun',
}

print(f"\n{'Token':<20}{'POS Tag':<10}{'Description'}")
print("-" * 60)
for token, tag in pos_tags[:15]:
    desc = tag_descriptions.get(tag, tag)
    print(f"{token:<20}{tag:<10}{desc}")
```

## 9. Bag of Words (BoW) Representation

```python
print("=== BAG OF WORDS ===")

# Clean corpus
cleaned_corpus = [clean_text(doc) for doc in corpus]

# Create BoW
bow_vectorizer = CountVectorizer(stop_words='english')
bow_matrix = bow_vectorizer.fit_transform(cleaned_corpus)

print(f"\nVocabulary size: {len(bow_vectorizer.get_feature_names_out())}")
print(f"BoW Matrix shape: {bow_matrix.shape}")

# Display top words
word_freq = np.array(bow_matrix.sum(axis=0)).flatten()
vocab = bow_vectorizer.get_feature_names_out()
freq_df = pd.DataFrame({'word': vocab, 'frequency': word_freq})
freq_df = freq_df.sort_values('frequency', ascending=False)

print(f"\nTop 15 words by total frequency across all documents:")
for _, row in freq_df.head(15).iterrows():
    print(f"{row['word']:<17}: {int(row['frequency'])}")
```

## 10. TF-IDF Vectorization

```python
print("=== TF-IDF VECTORIZATION ===")

tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(cleaned_corpus)

print(f"\nTF-IDF Matrix shape: {tfidf_matrix.shape}")

feature_names = tfidf_vectorizer.get_feature_names_out()

# Show top TF-IDF terms for each document
for doc_idx in range(2):  # Show for first 2 docs
    tfidf_scores = tfidf_matrix[doc_idx].toarray().flatten()
    top_indices = tfidf_scores.argsort()[::-1][:10]
    print(f"\nTop 10 TF-IDF terms for Document {doc_idx + 1}:")
    for idx in top_indices:
        print(f"  {feature_names[idx]:<19}: {tfidf_scores[idx]:.4f}")
```

## 11. Visualization — Word Frequency

```python
fig, ax = plt.subplots(figsize=(12, 5))

top_words = freq_df.head(15)
colors = plt.cm.viridis(np.linspace(0.2, 0.9, len(top_words)))

bars = ax.barh(top_words['word'], top_words['frequency'], color=colors)
ax.set_xlabel('Frequency', fontsize=12)
ax.set_title('Top 15 Most Frequent Words (after stop-word removal)', fontsize=14, fontweight='bold')
ax.invert_yaxis()

for bar, val in zip(bars, top_words['frequency']):
    ax.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
            f'{int(val)}', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

## 12. Sentiment Analysis

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

print("=== SENTIMENT ANALYSIS (VADER) ===")

sia = SentimentIntensityAnalyzer()

print("\n--- Analyzing Corpus Documents ---")
for i, doc in enumerate(corpus, 1):
    scores = sia.polarity_scores(doc)
    sentiment = 'Positive' if scores['compound'] >= 0.05 else ('Negative' if scores['compound'] <= -0.05 else 'Neutral')
    print(f"\nDocument {i} (first 60 chars): {doc[:60]}...")
    print(f"  Scores: neg={scores['neg']:.4f}, neu={scores['neu']:.4f}, pos={scores['pos']:.4f}, compound={scores['compound']:.4f}")
    print(f"  Sentiment: {sentiment}")

# Additional sample sentences
print("\n\n--- Analyzing Sample Sentences ---")
test_sentences = [
    "NLP is an amazing and revolutionary technology!",
    "The preprocessing results were terrible and disappointing.",
    "The algorithm processed the data.",
    "This model achieves excellent results with great accuracy!"
]

for sent in test_sentences:
    scores = sia.polarity_scores(sent)
    sentiment = 'Positive' if scores['compound'] >= 0.05 else ('Negative' if scores['compound'] <= -0.05 else 'Neutral')
    print(f"\n\"{sent}\"")
    print(f"  Scores: neg={scores['neg']:.4f}, neu={scores['neu']:.4f}, pos={scores['pos']:.4f}, compound={scores['compound']:.4f}")
    print(f"  Sentiment: {sentiment}")
```

## 13. Complete NLP Preprocessing Pipeline

```python
def nlp_preprocessing_pipeline(text):
    """Complete NLP preprocessing pipeline."""
    # 1. Lowercase
    text = text.lower()
    
    # 2. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 3. Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    # 4. Tokenize
    tokens = word_tokenize(text)
    
    # 5. Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if t not in stop_words]
    
    # 6. Lemmatize
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    
    return ' '.join(tokens)

print("=== COMPLETE NLP PREPROCESSING PIPELINE ===")
for i, doc in enumerate(corpus, 1):
    preprocessed = nlp_preprocessing_pipeline(doc)
    print(f"\n--- Document {i} ---")
    print(f"Original    : {doc}")
    print(f"Preprocessed: {preprocessed}")

print("\n✅ Pipeline complete!")
```

## 14. Conclusion

In this lab, we explored fundamental NLP concepts and applied essential text preprocessing techniques:

1. **Tokenization** — Breaking text into sentences and words (tokens)
2. **Text Cleaning** — Lowercasing, punctuation removal, whitespace normalization
3. **Stop-word Removal** — Filtering out common words that carry little meaning
4. **Stemming** — Reducing words to their root form (aggressive, may produce non-words)
5. **Lemmatization** — Reducing words to their dictionary form (more accurate than stemming)
6. **POS Tagging** — Identifying grammatical roles of words
7. **Bag of Words** — Representing text as word frequency vectors
8. **TF-IDF** — Weighting words by their importance across documents
9. **Sentiment Analysis** — Determining the emotional tone of text using VADER

**Key Takeaways:**
- Text preprocessing is essential before any NLP task — it standardizes and cleans raw text data.
- Lemmatization generally produces more meaningful results than stemming.
- TF-IDF is superior to simple BoW as it captures word importance relative to the entire corpus.
- VADER provides a simple, lexicon-based approach to sentiment analysis without requiring model training.
