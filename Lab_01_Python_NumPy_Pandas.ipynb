## 1. Import Libraries

```python
import numpy as np
import pandas as pd

print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
```

## 2. Python Programming Constructs

### 2.1 Variables, Data Types and Operators

```python
# Variables and data types
dataset_name = "Seeds Dataset"        # String
num_samples = 210                      # Integer
num_features = 7                       # Integer
avg_area = 14.85                       # Float
is_labeled = True                      # Boolean

print(f"Dataset: {dataset_name}")
print(f"Samples: {num_samples}, Features: {num_features}")
print(f"Average Area: {avg_area}")
print(f"Is Labeled: {is_labeled}")
print(f"Type of num_samples: {type(num_samples)}")
print(f"Type of avg_area: {type(avg_area)}")
print(f"Type of dataset_name: {type(dataset_name)}")

# Arithmetic Operators
print("\n--- Arithmetic Operators ---")
print(f"Total data points: {num_samples * num_features}")
print(f"Samples per class (approx): {num_samples // 3}")
print(f"Remainder: {num_samples % 3}")
```

### 2.2 Control Flow (if-else, loops)

```python
# If-Else
if num_samples > 200:
    print("Large dataset")
elif num_samples > 100:
    print("Medium dataset")
else:
    print("Small dataset")

# For loop
features = ['Area', 'Perimeter', 'Compactness', 'Length_of_kernel',
             'Width_of_kernel', 'Asymmetry_coefficient', 'Length_of_kernel_groove']

print("\nFeatures in the dataset:")
for i, feature in enumerate(features, 1):
    print(f"  {i}. {feature}")

# While loop
print("\nPowers of 2 up to num_features:")
power = 0
while 2**power <= num_features:
    print(f"  2^{power} = {2**power}")
    power += 1
```

### 2.3 Functions

```python
def calculate_stats(data):
    """Calculate basic statistics for a list of numbers."""
    return {
        'mean': sum(data) / len(data),
        'min': min(data),
        'max': max(data),
        'range': max(data) - min(data),
        'count': len(data)
    }

sample_areas = [15.26, 14.88, 14.29, 13.84, 16.14, 14.38, 14.69]
stats = calculate_stats(sample_areas)
print("Statistics for sample areas:")
for key, value in stats.items():
    print(f"  {key}: {value:.4f}")

# Lambda function
normalize = lambda x, min_val, max_val: (x - min_val) / (max_val - min_val)
print(f"\nNormalized 15.26: {normalize(15.26, stats['min'], stats['max']):.4f}")
```

### 2.4 List Comprehensions and Collections

```python
# List comprehension
squared_areas = [x**2 for x in sample_areas]
print(f"Original areas: {sample_areas}")
print(f"Squared areas:  {[round(x, 2) for x in squared_areas]}")

# Dictionary comprehension
feature_lengths = {f: len(f) for f in features}
print(f"\nFeature name lengths: {feature_lengths}")

# Tuple
dataset_shape = (num_samples, num_features)
print(f"\nDataset shape (tuple): {dataset_shape}")

# Set
classes = {1, 2, 3, 1, 2, 3}
print(f"Unique classes (set): {classes}")
```

## 3. NumPy Operations

### 3.1 Array Creation and Properties

```python
# Create arrays
arr_1d = np.array(sample_areas)
arr_2d = np.array([[15.26, 14.84, 0.871], [14.88, 14.57, 0.881], [14.29, 14.09, 0.905]])

print("1D Array:", arr_1d)
print(f"  Shape: {arr_1d.shape}, Dtype: {arr_1d.dtype}, Ndim: {arr_1d.ndim}")

print("\n2D Array:\n", arr_2d)
print(f"  Shape: {arr_2d.shape}, Dtype: {arr_2d.dtype}, Ndim: {arr_2d.ndim}")

# Special arrays
print("\nZeros (3x3):\n", np.zeros((3, 3)))
print("\nOnes (2x4):\n", np.ones((2, 4)))
print("\nIdentity (3x3):\n", np.eye(3))
print("\nLinspace (0 to 1, 5 pts):", np.linspace(0, 1, 5))
print("\nArange (0 to 10, step 2):", np.arange(0, 10, 2))
```

### 3.2 Array Operations

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])

# Element-wise operations
print("a + b:", a + b)
print("a * b:", a * b)
print("a ** 2:", a ** 2)
print("sqrt(b):", np.sqrt(b))

# Statistical operations
print(f"\nMean of arr_1d: {np.mean(arr_1d):.4f}")
print(f"Std of arr_1d: {np.std(arr_1d):.4f}")
print(f"Median of arr_1d: {np.median(arr_1d):.4f}")
print(f"Sum of arr_1d: {np.sum(arr_1d):.4f}")

# Matrix operations
mat_a = np.array([[1, 2], [3, 4]])
mat_b = np.array([[5, 6], [7, 8]])
print(f"\nMatrix multiplication:\n{np.dot(mat_a, mat_b)}")
print(f"\nTranspose of mat_a:\n{mat_a.T}")
```

### 3.3 Indexing, Slicing, and Boolean Masking

```python
# Indexing and slicing
print("arr_1d:", arr_1d)
print("First 3 elements:", arr_1d[:3])
print("Last 2 elements:", arr_1d[-2:])
print("Every other element:", arr_1d[::2])

# Boolean masking
mask = arr_1d > 14.5
print(f"\nElements > 14.5: {arr_1d[mask]}")
print(f"Count > 14.5: {np.sum(mask)}")

# Where
labels = np.where(arr_1d > 15, 'High', 'Low')
print(f"\nLabels: {labels}")
```

## 4. Pandas Operations

### 4.1 Loading and Exploring the Dataset

```python
# Load the Seeds dataset
df = pd.read_csv('seeds_dataset.csv')

print("Shape of dataset:", df.shape)
print(f"\nFirst 5 rows:")
df.head()
```

```python
# Dataset info
print("Dataset Info:")
df.info()
```

```python
# Data types
print("Data Types:")
print(df.dtypes)

print(f"\nColumn names: {list(df.columns)}")
print(f"Index: {df.index}")
```

### 4.2 Basic Statistics

```python
# Descriptive statistics
df.describe()
```

```python
# Class distribution
print("Class Distribution:")
print(df['Class'].value_counts())
print(f"\nUnique classes: {df['Class'].unique()}")
print(f"Number of unique classes: {df['Class'].nunique()}")
```

### 4.3 Data Selection and Filtering

```python
# Select specific columns
print("Area and Perimeter columns:")
print(df[['Area', 'Perimeter']].head())

# Filter rows
print("\nSamples with Area > 18:")
large_seeds = df[df['Area'] > 18]
print(f"Count: {len(large_seeds)}")
print(large_seeds.head())

# Multiple conditions
print("\nClass 1 seeds with Area > 15:")
filtered = df[(df['Class'] == 1) & (df['Area'] > 15)]
print(f"Count: {len(filtered)}")
print(filtered.head())
```

### 4.4 GroupBy and Aggregation

```python
# Group by Class
print("Mean of each feature by class:")
df.groupby('Class').mean()
```

```python
# Multiple aggregations
print("Area statistics by class:")
df.groupby('Class')['Area'].agg(['mean', 'std', 'min', 'max'])
```

### 4.5 Data Manipulation

```python
# Add a new column
df['Area_Category'] = pd.cut(df['Area'], bins=3, labels=['Small', 'Medium', 'Large'])
print("Area Category distribution:")
print(df['Area_Category'].value_counts())

# Sorting
print("\nTop 5 by Area:")
print(df.nlargest(5, 'Area')[['Area', 'Perimeter', 'Class']])

# Drop the added column
df.drop('Area_Category', axis=1, inplace=True)
```

### 4.6 Handling Missing Values

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")

# Demonstrate handling (create a copy with artificial NaN)
df_copy = df.copy()
df_copy.loc[0, 'Area'] = np.nan
df_copy.loc[5, 'Perimeter'] = np.nan

print("\nAfter introducing NaN:")
print(df_copy.isnull().sum())

# Fill NaN with mean
df_filled = df_copy.fillna(df_copy.mean(numeric_only=True))
print("\nAfter filling NaN with mean:")
print(df_filled.isnull().sum())
```

### 4.7 Correlation Analysis

```python
# Correlation matrix
print("Correlation Matrix:")
corr_matrix = df.select_dtypes(include=[np.number]).corr()
corr_matrix
```
